## 📢 1.1 What is Programming?
**📝 Explanation:**
Programming is the process of giving instructions to a computer.
Computers don’t understand human language, so we write code in programming languages.
A program is a set of instructions that tells the computer what to do.

**🔹 Real-Life Analogy:**
Think of programming like giving directions to a delivery person. If you don’t give clear and correct instructions, they might get lost! Similarly, in programming, clear and precise instructions (code) are needed for the computer to execute tasks.

### 📌 Key Takeaways:
- ✔ Computers follow exact instructions.
- ✔ Programming is how we tell computers what to do.
- ✔ Python is one of the easiest and most popular languages to start with.

## 📢 1.2 Why Python?
**✅ Why Choose Python?**
- ✔ Easy to Learn: Python’s syntax is simple and readable.
- ✔ Widely Used: Used in web development, AI, data science, automation, etc.
- ✔ Beginner-Friendly: Great for first-time programmers.

| Speed of Learning | ✅ Fast | ❌ Medium | ❌ Slow |

**🔹 Fun Fact:**
Python is used by companies like Google, Instagram, Netflix, and NASA! 🚀

## 📢 1.3 Setting Up Python
**🔹 Step 1: Install Python**
1️⃣ Go to [python.org](https://www.python.org/downloads/)
2️⃣ Download the latest version for Windows/Mac/Linux
3️⃣ Run the installer and check **"Add Python to PATH"** before installing

**🔹 Step 2: Install a Code Editor**
- 🔹 **Option 1**: Install [VS Code](https://code.visualstudio.com/) (Recommended)
- 🔹 **Option 2**: Use Python’s built-in IDLE (comes with Python)

**🔹 Step 3: Verify Installation**
✅ Open a terminal (Command Prompt or Terminal)
✅ Type `python --version` → Should display something like `Python 3.11.2`

## 📢 1.4 Writing Your First Python Program

```
print("Hello, World!")
```

**🔹 What Happens?**
When you run this code, Python prints **"Hello, World!"** on the screen.

**📝 Explanation:**
- `print()` is a built-in function that displays text on the screen.
- "Hello, World!" is a **string** (a sequence of characters enclosed in quotes).

**🎯 Mini Challenge:**
Try changing **"Hello, World!"** to **"Hello, Python!"** and run the code again.
